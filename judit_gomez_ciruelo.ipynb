```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
def precio_compra(articulos,compra):
    precios = pd.Series(articulos)
    factura_compra = precios[compra]
    total = reduce(lambda a,b : a+b,factura_compra)
    return total
price=precio_compra(articulos,compra)
price
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%
discount = 10

def precio_con_descuento (precio_compra,discount):
    descuento = precio_compra*discount/100
    total_with_discount = precio_compra-descuento
    return total_with_discount
precio_con_descuento(price,discount)
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
def superior_100 (precio_compra,discount):
    if precio_compra > 100:
        descuento = precio_compra*discount/100
        total_with_discount = precio_compra-descuento
        return total_with_discount
    else:
        return precio_compra

superior_100(price,discount)  
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)
iva=21

def precio_con_iva (precio_compra,iva):
    impuesto = precio_compra*iva/100
    total_with_iva = precio_compra+impuesto
    return total_with_iva

precio_con_iva(price,iva)
```

```python
def notas(curso, alumnos):
  total = []
  for i in range(len(alumnos)):
    if alumnos[i][2] == curso:
      nombre = (alumnos[i][0]+ " " +alumnos[i][1])
      promedio = (sum(alumnos[i][3])/len(alumnos[i][3]))
      total.append((nombre,promedio))
  return total
```
